## Correlations

```python
# To create a subset of the frame with categories for correlation testing and remove outliers
df_subset=df[['2015_HB_2778_Eligible?','2019_HB_2042_Eligible?','Sale_Price','Electric_Range','Base_MSRP','Odometer_Reading']]
```

```python
df_sub2=df_subset[(df_subset['Sale_Price']!=0)&(df_subset['Sale_Price']<=539130)&(df_subset['Electric_Range']!=0)&(df_subset['Base_MSRP']!=0)&(df_subset['Base_MSRP']<=126136)&(df_subset['Odometer_Reading']!=0)&(df_subset['Odometer_Reading']<=120800)]
```

```python
df_sub2.shape
```

```python
# To isolate records of vehicles that were eligible for either the 2015 or 2019 clean alternative fuel vehicle tax exemptions
df_eligible=df_sub2[(df_sub2['2015_HB_2778_Eligible?']=='Eligible') | (df_sub2['2019_HB_2042_Eligible?']=='Eligible')]
```

```python
df_eligible.head(20)
```

```python
df_eligible.shape
```

##### df_eligible allows me to look for correlations between variables among EVs which qualified for either the 2015 or the 2019 tax exemption while keeping the rest of my data intact in df.

```python
# To create a correlation matrix
df_eligible.corr()
```

```python
# To create a subplot with matplotlib and to create a heatmap with seaborn based on this subplot
f,ax=plt.subplots(figsize=(10,10))
corr=sns.heatmap(df_eligible.corr(),annot=True,ax=ax)
```

```python
df_sub2.corr()
```

```python
# To create a subplot and a heatmap with based on df_sub2
f,ax=plt.subplots(figsize=(10,10))
corr=sns.heatmap(df_sub2.corr(),annot=True,ax=ax)
```

##### The first heat map displays correlations among variables for EVs eligible for at least one tax exemption, while the second heat map displays correlations among variables for all EVs titled and registered in Washington state from 2010 to 2022. Variables for exemption eligible EVs tended to have much stronger correlations, possibly due to a lack of variety in eligible vehicle models.

##### The strong positive relationship between odometer reading and electric range in both the eligible set and the general set indicate that people drive more if their EVs have a higher electric range.

##### There was a strong positive correlation between electric range and base MSRP, which indicates that cars with a higher electric range are more valuable. However, this relationship and the relationship between electric range and odometer reading may be confounding the correlation between odometer reading and base MSRP. One would expect the value of a car to decrease with increased miles on the odometer, however there appears to be a positive correlation between the two variables.

##### The correlation between sale price and odometer reading is negative, which is to be expected, but the relationship is weak in both sets.

## Scatterplots

```python
# To create a scatterplot for the odometer_rating and electric_range columns in the exemption eligible set
sns.lmplot(x='Odometer_Reading', y='Electric_Range', data = df_eligible)
```

##### Electric range seems to fall into categories at around 20, 100, 120, 200, and 275 miles, probably due to what kind of bateries they are using. The same kinds of battery perform very similarly in EV range tests. Vehicles with a 200 mile range seem to have the most miles on their odometers, while vehicles with a range of around 275 have less, departing from the strong positive trend slightly. This might be because vehicles with the highest electric range are newer on the market and haven't had the time to put as many miles on their odometers.

```python
# To create a scatter plot for the Base_MSRP and Odometer_Reading columns
sns.lmplot(x='Base_MSRP', y='Electric_Range', data = df_eligible)
```

##### Here we see the same five electric range categories and far fewer data points, showing us that there are not many EV models on the market which were eligible for Washington's tax exemptions. The models that had the lowest electric range varied considerably in their base MSRP, possibly because they were hybrid electric vehicles that didn't rely 100% on electricity (accounting for the low electric range) and they ranged from economy class to more luxury vehicles (accounting for the range in value). It is important to note that of the EVs which were eligible for Washington's tax exemptions, only 9 had electric range, sale price, odometer reading, and base MSRP data available. It is likely that as more EVs enter the market and more data becomes available we will see a much weaker correlation similar to that present in the general set.

## Categorical plots

```python
# To create a categorical plot for Sale_Price
sales=df_sub2[(df_sub2['Sale_Price']<=100000)]
sns.histplot(sales['Sale_Price'],bins=20,kde=True)
```

```python
sales.loc[sales['Sale_Price']<30000,'Price_Category']='Low Price'
sales.loc[(sales['Sale_Price']>=30000)&(sales['Sale_Price']<50000),'Price_Category']= 'Middle Price'
sales.loc[sales['Sale_Price']>=50000,'Price_Category']='High Price'
```

```python
sales['Price_Category'].value_counts()
```

```python
sns.set(style='ticks')
g=sns.catplot(x='Electric_Range', y='Sale_Price', hue='Price_Category',data=sales)
```

##### Of the vehicles in the general set, the 12 to 19 mile range and the 208+ mile range have the most high price sales. Middle price vehicles are evenly distributed across electric ranges, and low price vehicles are most common in the 12 to 14 mile and 93+ mile electric range.

##### One of the key questions this project seeks to answer is 'what are people willing to pay for a clean alternative fuel vehicle?' This histogram suggests that most people who registered their clean alternative fuel vehicle with the Washington state DOL would be willing to pay up to 40,000 USD for a clean alternative fuel vehicle, though there is also a sizeable group (perhaps those buying second hand vehicles) who would be willing to pay 15,000 USD

```python
sns.histplot(df_sub2['Electric_Range'],bins=20,kde=True)
```

```python
df_sub2.loc[df_sub2['Electric_Range']<50,'Range_Category']='Short range'
df_sub2.loc[(df_sub2['Electric_Range']>=50)&(df_sub2['Electric_Range']<200),'Range_Category']= 'Medium range'
df_sub2.loc[df_sub2['Electric_Range']>=200,'Range_Category']='Long range'
```

## Hypothesis

##### I hypothesize that as electric range continues to improve and get longer, the base MSRP value of these long-range vehicles will get higher.

##### This is an important relationship to keep an eye on because if it holds true, the average buyer who is willing to pay up to 40,000 USD for an EV might be priced out of buying a long range EV. This would run counter to the goals of Washington state, as the state wants to encourage more EV purchases.
